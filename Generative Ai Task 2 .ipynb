```python
import ollama

# Change this to whichever model you pulled
MODEL_NAME = "llama3.2"
```

```python
def generate_response(prompt, model=MODEL_NAME):
    try:
        response = ollama.chat(
            model=model,
            messages=[{"role": "user", "content": prompt}]
        )
        return response['message']['content']
    except Exception as e:
        print(f"ERROR for prompt '{prompt[:40]}...': {type(e).__name__}: {e}")
        return None
```

```python
user_prompt = input("Enter your prompt: ")
output = generate_response(user_prompt)

if output:
    print("\n--- Generated Response ---\n")
    print(output)
```

```python
test_prompts = [
    "Explain what a Large Language Model is in simple terms.",
    "Write a short poem about the ocean.",
    "What is the difference between AI and Machine Learning?",
    "Summarize the plot of Cinderella in 2 sentences."
]

results = {}

for i, prompt in enumerate(test_prompts, 1):
    print(f"\n===== Test {i}: {prompt} =====")
    response = generate_response(prompt)
    results[prompt] = response
    print(response)
    print("-" * 60)
```

```python
models_to_compare = ["llama3.2", "phi3:mini"]  # pull both first via `ollama pull`
prompt = "Explain photosynthesis in one paragraph."

for model in models_to_compare:
    print(f"\n=== Model: {model} ===")
    print(generate_response(prompt, model=model))
    print("-" * 60)
```

```python
import pandas as pd

df = pd.DataFrame(list(results.items()), columns=["Prompt", "Response"])
df
```
